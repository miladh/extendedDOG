Constant tests:
-----

```python
def constant(ns, nt, ds, dt, c, id):
    print "TEST(constant_" + str(id) + "){"
    
    print "int ns = " + str(ns) + ";"
    print "int nt = " + str(nt) + ";"
    print "double ds = " + str(ds) + ";"
    print "double dt = " + str(dt) + ";\n"
    
    print "int Ns = pow(2,ns);"
    print "int Nt = pow(2,nt);\n"
    
    print "double c = " + str(c) + ";\n"
    
    print "Integrator integrator(nt, dt, ns, ds);\n"
    
    print "vec k = integrator.spatialFreqVec();"
    print "vec w = integrator.temporalFreqVec();\n"

    print "cx_cube g = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube G = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube f = zeros<cx_cube>(Ns, Ns, Nt);\n"
    
    print 
    print "//Spatiotemporal signal\n"\
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      g(i,j,l) = c; \n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    print "//fourier signal \n" \
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      f(i,j,l) = Functions::delta(k[i], 0)\n"\
    "               * Functions::delta(k[j], 0)\n"\
    "               * Functions::delta(w[l], 0);\n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    
    print "f *= c*8*PI*PI*PI;\n"\
    "f /= integrator.spatialFreqResolution()\n"\
    "* integrator.spatialFreqResolution()\n"\
    "* integrator.temporalFreqResolution();\n\n"\
    "// Backward \n"\
    "G = integrator.integrate(f);\n"\
    "G = FFTHelper::fftShift(G);\n\n"\
    "// Test \n"\
    "for(int l = 0; l < Nt; l++){\n"\
    "   for(int i = 0; i < Ns; i++){\n"\
    "      for(int j = 0; j < Ns; j++){\n"\
    "           CHECK_CLOSE(real(g(i,j,l)), \n"\
    "                       real(G(i,j,l)), 1e-9); \n"\
    "      }\n"\
    "   }\n"\
    "}"

    
    
    print"}"
    
#############################################################################
cVec = [-60.5 , -0.002, 0.0, 4.6, 32.1, 1000.5]
nsVec = range(1,3)
ntVec = range(1,3)

dsVec = linspace(0.01, 0.5, 3)
dtVec = linspace(0.01, 0.5, 3)

id = 0
for c in cVec:
    for ns in nsVec:
        for nt in ntVec:
            for ds in dsVec:
                for dt in dtVec:
                    constant(ns, nt, ds, dt, c, id)
                    id+=1
```

Cosine tests:
-----

```python
def cosine(ns, nt, ds, dt, wdId, kxId, kyId, id):
    print "TEST(cosine_" + str(id) + "){"
    
    print "int ns = " + str(ns) + ";"
    print "int nt = " + str(nt) + ";"
    print "double ds = " + str(ds) + ";"
    print "double dt = " + str(dt) + ";\n"
    
    print "int Ns = pow(2,ns);"
    print "int Nt = pow(2,nt);\n"
    
    print "Integrator integrator(nt, dt, ns, ds);\n"
    
    print "vec s = integrator.coordinateVec();"
    print "vec k = integrator.spatialFreqVec();"
    print "vec t = integrator.timeVec();"
    print "vec w = integrator.temporalFreqVec();\n"

    print "cx_cube g = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube G = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube f = zeros<cx_cube>(Ns, Ns, Nt);\n"
    
    print "double wd = w(" + str(wdId) + ");"
    print "double kx = k(" + str(kxId) + ");"
    print "double ky = k(" + str(kyId) + ");"
    
    print 
    print "//Spatiotemporal signal\n"\
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      g(i,j,l) =cos(kx*s[i] + ky*s[j] - wd * t[l]); \n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    print "//fourier signal \n" \
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      f(i,j,l) = Functions::delta(k[i], kx)\n"\
    "               * Functions::delta(k[j], ky)\n"\
    "               * Functions::delta(w[l], -wd);\n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    
    print "f *= 8*PI*PI*PI;\n"\
    "f /= integrator.spatialFreqResolution()\n"\
    "* integrator.spatialFreqResolution()\n"\
    "* integrator.temporalFreqResolution();\n\n"\
    "// Backward \n"\
    "G = integrator.integrate(f);\n"\
    "G = FFTHelper::fftShift(G);\n\n"\
    "// Test \n"\
    "for(int l = 0; l < Nt; l++){\n"\
    "   for(int i = 0; i < Ns; i++){\n"\
    "      for(int j = 0; j < Ns; j++){\n"\
    "           CHECK_CLOSE(real(g(i,j,l)), \n"\
    "                       real(G(i,j,l)), 1e-9); \n"\
    "      }\n"\
    "   }\n"\
    "}"

    
    print"}"
    
#############################################################################
nsVec = range(2,4)
ntVec = range(2,4)

dsVec = linspace(0.01, 0.5, 2)
dtVec = linspace(0.01, 0.5, 2)

id = 0
for ns in nsVec:
    for nt in ntVec:
        wdVec = range(0, 2**nt/2) + range(2**nt/2+1, 2**nt) 
        kxVec = range(1, 2**ns/2) + range(2**ns/2+1, 2**ns) 
        kyVec = range(1, 2**ns/2) + range(2**ns/2+1, 2**ns) 
        for ds in dsVec:
            for dt in dtVec:
                for wdId in wdVec:
                    for kxId in kxVec:
                        for kyId in kyVec:
                            cosine(ns, nt, ds, dt, wdId, kxId, kyId, id)
                            id+=1
```

Gauss tests:

```python
def gauss(ns, nt, ds, dt, a, wdId, id):
    print "TEST(gauss_" + str(id) + "){"
    
    print "int ns = " + str(ns) + ";"
    print "int nt = " + str(nt) + ";"
    print "double ds = " + str(ds) + ";"
    print "double dt = " + str(dt) + ";\n"
    
    print "int Ns = pow(2,ns);"
    print "int Nt = pow(2,nt);\n"
    
    print "Integrator integrator(nt, dt, ns, ds);\n"
    
    print "vec s = integrator.coordinateVec();"
    print "vec k = integrator.spatialFreqVec();"
    print "vec t = integrator.timeVec();"
    print "vec w = integrator.temporalFreqVec();\n"

    print "cx_cube g = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube G = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube f = zeros<cx_cube>(Ns, Ns, Nt);\n"
    
    print "double wd = w(" + str(wdId) + ");"
    print "double a =" + str(a) + ";"
    
    print 
    print "//Spatiotemporal signal\n"\
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      g(i,j,l) =gauss(a, {s[i], s[j]}) * cos(wd * t[l]); \n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    print "//fourier signal \n" \
    "for(int l = 0; l < Nt; l++){\n" \
    "  for(int i = 0; i < Ns; i++){\n"\
    "    for(int j = 0; j < Ns; j++){\n"\
    "      f(i,j,l) = gaussFT(a, {k[i], k[j]})\n"\
    "               * Functions::delta(wd, w[l]);\n"\
    "    }\n"\
    "  }\n"\
    "}\n"
    
    
    print "f *= 2*PI;\n"\
    "f /= integrator.temporalFreqResolution();\n\n"\
    "// Backward \n"\
    "G = integrator.integrate(f);\n"\
    "G = FFTHelper::fftShift(G);\n\n"\
    "// Test \n"\
    "for(int l = 0; l < Nt; l++){\n"\
    "   for(int i = 0; i < Ns; i++){\n"\
    "      for(int j = 0; j < Ns; j++){\n"\
    "           CHECK_CLOSE(real(g(i,j,l)), \n"\
    "                       real(G(i,j,l)), 1e-9); \n"\
    "      }\n"\
    "   }\n"\
    "}"

    
    print"}"
    
#############################################################################
nsVec = [9, 8, 8, 8]
aVec = [0.1, 2.2,5.8, 60.8]
dsVec = [0.1, 0.1, 0.1, 0.01]

nt= 2
dt = 0.1

id = 0

for i in range(len(aVec)):
    gauss(nsVec[i], nt, dsVec[i], dt, aVec[i], 2**nt-2, id)
    id+=1
```

```python
def grating(ns, nt, ds, dt, C, wdId, kxId, kyId, id):
    print "TEST(grating_" + str(id) + "){"
    
    print "int ns = " + str(ns) + ";"
    print "int nt = " + str(nt) + ";"
    print "double ds = " + str(ds) + ";"
    print "double dt = " + str(dt) + ";\n"
    
    print "int Ns = pow(2,ns);"
    print "int Nt = pow(2,nt);\n"
    
    print "Integrator integrator(nt, dt, ns, ds);\n"
    
    print "vec s = integrator.coordinateVec();"
    print "vec k = integrator.spatialFreqVec();"
    print "vec t = integrator.timeVec();"
    print "vec w = integrator.temporalFreqVec();\n"

    print "cx_cube g = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube G = zeros<cx_cube>(Ns, Ns, Nt);"
    print "cx_cube f = zeros<cx_cube>(Ns, Ns, Nt);\n"
    
    print "double C =" + str(C) + ";"
    print "double wd = w(" + str(wdId) + ");"
    print "double kx = k(" + str(kxId) + ");"
    print "double ky = k(" + str(kyId) + ");\n"
    
    print "Grating S(&integrator, {kx, ky}, wd, C);"

    
    print "//Spatiotemporal signal"
    print "S.computeSpatiotemporal();"
    print "g.set_real(S.spatioTemporal());"

    
    print "//fourier signal \n" 
    print "S.computeFourierTransform();"
    print "f = S.fourierTransform();"
    
    
    print "// Backward \n"\
    "G = integrator.integrate(f);\n"\
    "G = FFTHelper::fftShift(G);\n\n"\
    "// Test \n"\
    "for(int l = 0; l < Nt; l++){\n"\
    "   for(int i = 0; i < Ns; i++){\n"\
    "      for(int j = 0; j < Ns; j++){\n"\
    "           CHECK_CLOSE(real(g(i,j,l)), \n"\
    "                       real(G(i,j,l)), 1e-10); \n"\
    "      }\n"\
    "   }\n"\
    "}"

    
    print"}"
    
#############################################################################
nsVec = range(2,4)
ntVec = range(2,4)

CVec = [-0.1, 2.2]

dsVec = linspace(0.01, 0.5, 2)
dtVec = linspace(0.01, 0.5, 2)

id = 0
for ns in nsVec:
    for nt in ntVec:
        wdVec = range(0, 2**nt/2) + range(2**nt/2+1, 2**nt) 
        kxVec = range(1, 2**ns/2) + range(2**ns/2+1, 2**ns) 
        kyVec = range(1, 2**ns/2) + range(2**ns/2+1, 2**ns) 
        for ds in dsVec:
            for dt in dtVec:
                for wdId in wdVec:
                    for kxId in kxVec:
                        for kyId in kyVec:
                            for C in CVec:
                                grating(ns, nt, ds, dt, C, wdId, kxId, kyId, id)
                                id+=1
```
